# [dbo].[GenerarCodigo]

```sql
/****** Object:  StoredProcedure [dbo].[GenerarCodigo]    Script Date: 05-Nov-24 7:41:20 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
 CREATE PROCEDURE [dbo].[GenerarCodigo]
    @CodigoAlmacen VARCHAR(10),
    @NuevoCodigo VARCHAR(50) OUTPUT
AS
BEGIN
    DECLARE @AñoActual INT = YEAR(GETDATE());
    DECLARE @NumeroIncremental INT;

    -- Verificar si ya hay registros para el año actual y el código de almacén
    IF EXISTS (SELECT 1 FROM ControlIncremental WHERE Año = @AñoActual AND CodigoAlmacen = @CodigoAlmacen)
    BEGIN
        -- Obtener el último número para el año actual y el almacén específico
        SELECT @NumeroIncremental = Numero + 1 
        FROM ControlIncremental 
        WHERE Año = @AñoActual AND CodigoAlmacen = @CodigoAlmacen;

        -- Actualizar el número
        UPDATE ControlIncremental
        SET Numero = @NumeroIncremental
        WHERE Año = @AñoActual AND CodigoAlmacen = @CodigoAlmacen;
    END
    ELSE
    BEGIN
        -- Si es el primer registro del año para este almacén, inicializar el número
        SET @NumeroIncremental = 1;

        -- Insertar el nuevo registro en la tabla de control
        INSERT INTO ControlIncremental (Año, Numero, CodigoAlmacen)
        VALUES (@AñoActual, @NumeroIncremental, @CodigoAlmacen);
    END

    -- Generar el código en el formato deseado: "Incremental-Numero-Año-CodigoAlmacen"
    SET @NuevoCodigo = CAST(@NumeroIncremental AS VARCHAR(10)) + '-'
                      + RIGHT('0000' + CAST(@NumeroIncremental AS VARCHAR(10)), 4) + '-'
                      + CAST(@AñoActual AS VARCHAR(4)) + '-'
                      + @CodigoAlmacen;
END;
GO
```

# [dbo].[ObtenerProximoNumeroRequisicion]

```sql
/****** Object:  StoredProcedure [dbo].[ObtenerProximoNumeroRequisicion]    Script Date: 05-Nov-24 7:41:20 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[ObtenerProximoNumeroRequisicion]
    @NuevoNumero INT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;

    BEGIN TRY
        BEGIN TRANSACTION;

        -- Obtener el último número de requisición usado
        DECLARE @UltimoNumero INT;

        SELECT @UltimoNumero = ISNULL(MAX(CAST(Nrequisicion AS INT)), 0)
        FROM Transaccion WITH (UPDLOCK, HOLDLOCK); -- Bloquear la tabla para evitar lecturas sucias

        -- Incrementar el número de requisición
        SET @NuevoNumero = @UltimoNumero + 1;

        -- Confirmar la transacción si todo está correcto
        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        -- Si ocurre algún error, se hace rollback de la transacción
        IF @@TRANCOUNT > 0
        BEGIN
            ROLLBACK TRANSACTION;
        END

        -- Usar RAISERROR en lugar de THROW para versiones anteriores a SQL Server 2012
        DECLARE @ErrorMessage NVARCHAR(4000), @ErrorSeverity INT, @ErrorState INT;
        SELECT @ErrorMessage = ERROR_MESSAGE(), @ErrorSeverity = ERROR_SEVERITY(), @ErrorState = ERROR_STATE();
        RAISERROR (@ErrorMessage, @ErrorSeverity, @ErrorState);
    END CATCH
END;
GO
```
